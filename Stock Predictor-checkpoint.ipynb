```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
start = '20-01-01'
end = '2023-11-10'
stock = 'TSLA'

data = yf.download(stock, start, end)
```

```python
data.reset_index(inplace=True)
```

```python
data.head()
```

```python
plt.figure(figsize=(8,6))
plt.plot(data.Close,'r')
plt.show()
```

```python
# Calculate Simple Moving Average (SMA)
def calculate_sma(data, window_size):
    sma = data.rolling(window=window_size).mean()
    return sma

# Choose the window size for SMA
sma_window = 20  # Adjust the window size as needed
data['SMA'] = calculate_sma(data['Close'], sma_window)

# Plot the original data and SMA
plt.figure(figsize=(10, 6))
plt.plot(data['Close'], label='Original Data')
plt.plot(data['SMA'], label=f'SMA ({sma_window} days)', color='orange')
plt.title('Stock Price with Simple Moving Average')
plt.legend()
plt.show()
```

```python
# Calculate Exponential Moving Average (EMA)
def calculate_ema(data, alpha):
    ema = data.ewm(span=alpha, adjust=False).mean()
    return ema

# Choose the alpha value for EMA
ema_alpha = 20  # Adjust the alpha value as needed
data['EMA'] = calculate_ema(data['Close'], ema_alpha)

# Plot the original data and EMA
plt.figure(figsize=(10, 6))
plt.plot(data['Close'], label='Original Data')
plt.plot(data['EMA'], label=f'EMA ({ema_alpha} days)', color='green')
plt.title('Stock Price with Exponential Moving Average')
plt.legend()
plt.show()
```

```python
closeData = data.reset_index()['Close']
```

```python
closeData
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler(feature_range=(0,1))
closeData=scaler.fit_transform(np.array(closeData).reshape(-1,1))
```

```python
closeData
```

```python
training_size=int(len(closeData)*0.65)
test_size=len(closeData)-training_size
train_data,test_data=closeData[0:training_size,:],closeData[training_size:len(closeData),:1]
```

```python
training_size,test_size
```

```python
import numpy
# convert an array of values into a dataset matrix
def create_dataset(dataset, time_step=1):
	dataX, dataY = [], []
	for i in range(len(dataset)-time_step-1):
		a = dataset[i:(i+time_step), 0]   ###i=0, 0,1,2,3-----99   100 
		dataX.append(a)
		dataY.append(dataset[i + time_step, 0])
	return numpy.array(dataX), numpy.array(dataY)
```

```python
# reshape into X=t,t+1,t+2,t+3 and Y=t+4
time_step = 100
X_train, y_train = create_dataset(train_data, time_step)
X_test, y_test = create_dataset(test_data, time_step)
```

```python
print(X_train.shape), print(y_train.shape)
```

```python
print(X_test.shape), print(y_test.shape)
```

```python
# reshape input to be [samples, time steps, features] which is required for LSTM
X_train =X_train.reshape(X_train.shape[0],X_train.shape[1] , 1)
X_test = X_test.reshape(X_test.shape[0],X_test.shape[1] , 1)
```

```python
### Create the Stacked LSTM model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LSTM
```

```python
model=Sequential()
model.add(LSTM(50,return_sequences=True,input_shape=(100,1)))
model.add(LSTM(50,return_sequences=True))
model.add(LSTM(50))
model.add(Dense(1))
model.compile(loss='mean_squared_error',optimizer='adam')
```

```python
import tensorflow as tf
tf.__version__
```

```python
model.summary()
```

```python
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=50,batch_size=64,verbose=1)
```

```python
train_predict=model.predict(X_train)
test_predict=model.predict(X_test)
```

```python
train_predict=scaler.inverse_transform(train_predict)
test_predict=scaler.inverse_transform(test_predict)
```

```python
import math
from sklearn.metrics import mean_squared_error
math.sqrt(mean_squared_error(y_train,train_predict))
```

```python
### Test Data RMSE
math.sqrt(mean_squared_error(y_test,test_predict))
```

```python
### Plotting 
# shift train predictions for plotting
look_back=100
trainPredictPlot = numpy.empty_like(closeData)
trainPredictPlot[:, :] = np.nan
trainPredictPlot[look_back:len(train_predict)+look_back, :] = train_predict
# shift test predictions for plotting
testPredictPlot = numpy.empty_like(closeData)
testPredictPlot[:, :] = numpy.nan
testPredictPlot[len(train_predict)+(look_back*2)+1:len(closeData)-1, :] = test_predict
# plot baseline and predictions
plt.plot(scaler.inverse_transform(closeData))
plt.plot(trainPredictPlot)
plt.plot(testPredictPlot)
plt.show()
```
